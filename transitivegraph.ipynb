```python
import networkx as nx
"""
networkx's format for adjacency list is like this:

A A1 A2 A3 A4 ...
B B1 B2 B3 B4 ...
C B1 B2 B3 B4 ...
.
.
.

And this means that A is connected to A1,A2,...
Then B is connected to B1,B2,B3,...
Then C is connected to C1,C2,C3,...
"""


adj="""
0 0 1 3
1 1 2
2 2
3 3
"""

G = nx.parse_adjlist(adj.split("\n")[1:], nodetype=int, create_using=nx.DiGraph)
W = G.copy()

#nx.draw(G,with_labels = True)
#nx.draw_circular(G, with_labels=True)

# Warshall algorithm
for k in W.nodes:
  for i in W.nodes:
    for j in W.nodes:
      if W.has_edge(i,k) and W.has_edge(k,j):
        W.add_edge(i,j)

import matplotlib.pyplot as plt

plt.figure(1)
nx.draw_circular(G, with_labels=True)
plt.figure(2)
nx.draw_circular(W, with_labels=True)

print(G.nodes)
print(nx.adjacency_matrix(G,nodelist=sorted(W.nodes)).toarray())
print(nx.adjacency_matrix(W,nodelist=sorted(W.nodes)).toarray())alex anthology for algorithms
```
